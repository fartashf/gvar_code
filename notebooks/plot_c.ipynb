```python
import numpy as np
import pylab as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
plt.rcParams.update({'font.size': 16})
```

# uniform 2D data

```python
N = 1000
D = 2
X = np.random.uniform(-1, 1, (D, N))
# X = np.random.randn(D, N)
```

```python
plt.scatter(X[0], X[1])
```

```python
num_cs = 2**np.arange(9)
dists = np.zeros(len(num_cs))
for i, nc in enumerate(num_cs):
    kmeans = KMeans(n_clusters=nc, random_state=0).fit(X.T)
    C = kmeans.cluster_centers_.T
    plt.figure()
    plt.ion()
    plt.scatter(X[0], X[1], label='X')
    plt.scatter(C[0], C[1], label='C')
    plt.legend()
    dists[i] = kmeans.inertia_  # -(X**2).sum()
```

```python
plt.plot(num_cs, dists, label='kmeans', c='r')
plt.plot(num_cs, dists[0]/np.power(num_cs, 1/D), label=r'$D_0/C^{1/d}$', c='k')
plt.plot(num_cs, dists[0]/np.sqrt(num_cs), label=r'$D_0/\sqrt{C}$', c='b')
plt.plot(num_cs, dists[0]/num_cs, label=r'$D_0/C$', c='b', ls='--')
plt.plot(num_cs, dists[0]/(2*num_cs), label=r'$D_0/2C$', c='b', ls='-.')
plt.plot(num_cs, dists[0]/(num_cs**2), label=r'$1/C^2$', c='b', ls=':')
plt.gca().set_yscale('log')
plt.legend()
```

# Mixture of gaussians

```python
K = 20
N = 1000
D = 2
a_gmm = 10
C_gmm = []
X = []
for k in range(K):
    C_gmm += [np.random.uniform(-a_gmm, a_gmm, (D, 1))]
    X += [np.random.randn(D, N//K)+C_gmm[-1]]
X = np.hstack(X)
```

```python
plt.scatter(X[0], X[1])
```

```python
num_cs = 2**np.arange(9)
dists = np.zeros_like(num_cs)
for i, nc in enumerate(num_cs):
    kmeans = KMeans(n_clusters=nc, random_state=0).fit(X.T)
    C = kmeans.cluster_centers_.T
    plt.figure()
    plt.ion()
    plt.scatter(X[0], X[1], label='X')
    plt.scatter(C[0], C[1], label='C')
    plt.legend()
    dists[i] = kmeans.inertia_
```

```python
plt.plot(num_cs, dists, label='kmeans', c='r')
plt.plot(num_cs, dists[0]/np.power(num_cs, 1/D), label=r'$D_0/C^{1/d}$', c='k')
plt.plot(num_cs, dists[0]/np.sqrt(num_cs), label=r'$D_0/\sqrt{C}$', c='b')
plt.plot(num_cs, dists[0]/num_cs, label=r'$D_0/C$', c='b', ls='--')
plt.plot(num_cs, dists[0]/(2*num_cs), label=r'$D_0/2C$', c='b', ls='-.')
plt.plot(num_cs, dists[0]/(num_cs**2), label=r'$1/C^2$', c='b', ls=':')
plt.gca().set_yscale('log')
plt.legend()
```

- from GMM 2D, there should be a level of separabality and mixing that the curve drops to D0/C but stays smooth after the turning point
- agmm is important then
- from uniform and gaussian 1000D, kmeans is at the top matching D0/C^1/d
- centers from normal or uniform doesn't change the shape of the curve, we need a level of mixing not provided by GMM
- So we know that gluster is doing much better than clustering on the uniform data
