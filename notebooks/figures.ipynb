```python
fnames = [
'20_dataset_mnist,optim_dmom,lr_0.01,dmom_0.0,momentum_0.9,alpha_normg,alpha_norm_exp,norm_temp_1,sampler_alpha_perc_90',
'24_dataset_mnist,optim_dmom,lr_0.01,dmom_0.0,momentum_0.9,alpha_ggbar_abs,alpha_norm_exp,norm_temp_0.01,sampler_alpha_perc_90',
]
```

```python
data=torch.load(prefix+fnames[0]+'/model.pth.tar')
opt = DictWrapper(data['opt'])
opt.cuda = True
train_loader, test_loader = get_loaders(opt)
```

```python
plt.rcParams.update({'font.size': 12})
names = ['DBo-$\|g_i\|$', 'DBo-$\|g_i^Tg\|$']
for n in range(len(fnames)):
    data=torch.load(prefix+fnames[n]+'/model.pth.tar')
    alphas = data['alpha_normed']
    big_alphas = np.argsort(alphas)[-16:][::-1]
    small_alphas = np.argsort(alphas)[:16]
    x = [train_loader.dataset[i][0] for i in big_alphas]
    x = vutils.make_grid(x, nrow=16, normalize=True, scale_each=True)
    plt.figure()
    plt.imshow(x.cpu().numpy().transpose([1,2,0]))
    plt.axis('off')
    # plt.title('Big alphas for %s' % names[n])
    plt.savefig(prefix+'fig%d_big.png'% n, dpi=100, bbox_inches='tight')
    x = [train_loader.dataset[i][0] for i in small_alphas]
    x = vutils.make_grid(x, nrow=16, normalize=True, scale_each=True)
    plt.figure()
    plt.imshow(x.cpu().numpy().transpose([1,2,0]))
    plt.axis('off')
    # plt.title('Small alphas for %s' % names[n])
    plt.savefig(prefix+'fig%d_small.png' % n, dpi=100, bbox_inches='tight')
```

```python
names = ['DBo-$\|g_i\|$', 'DBo-$\|g_i^Tg\|$']
for n in range(len(fnames)):
    plt.figure()
    plt.rcParams.update({'font.size': 12})
    plt.grid(linewidth=1)
    data=torch.load(prefix+fnames[n]+'/model.pth.tar')
    alphas = data['alpha_normed']
    plt.hist(np.log(alphas[alphas>np.exp(-20)]), bins=100)
    plt.ylim([0, 500])
    plt.xlim([-20, 0])
    plt.title('# Total non-zeros: %d/60000' % (alphas>np.exp(-20)).sum())
    plt.xlabel('$\log(alpha)$')
    plt.ylabel('# Training Examples')
    plt.savefig(prefix+'hist%d.png'%n, dpi=100, bbox_inches='tight')
```
